```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Exercício 1: Probabilidade de dois lançamentos de moedas carregadas

# Probabilidades das moedas carregadas
prob_table = pd.DataFrame({
    'p1(H)': [0.5, 0.6, 0.4, 0.1, 0.3],
    'p2(H)': [0.5, 0.7, 0.8, 0.2, 0.4]
})

# Funções auxiliares para calcular as probabilidades
prob_table['H-H'] = prob_table['p1(H)'] * prob_table['p2(H)']
prob_table['H-T'] = prob_table['p1(H)'] * (1 - prob_table['p2(H)'])
prob_table['T-H'] = (1 - prob_table['p1(H)']) * prob_table['p2(H)']
prob_table['T-T'] = (1 - prob_table['p1(H)']) * (1 - prob_table['p2(H)'])

# Mostrar a tabela com todas as probabilidades calculadas
prob_table
```

```python
# Exercício 2: Filtro de Spam e o Teorema de Bayes

# Probabilidades dadas
P_spam = 0.6
P_non_spam = 0.4
P_pos_given_spam = 0.98
P_pos_given_non_spam = 0.02

# Teorema de Bayes para calcular P(non_spam | Pos)
P_pos = P_pos_given_spam * P_spam + P_pos_given_non_spam * P_non_spam
P_non_spam_given_pos = (P_pos_given_non_spam * P_non_spam) / P_pos

# Número de e-mails sinalizados como spam
n_spam = 963

# E-mails suspeitos que são não-spam
non_spam_emails = P_non_spam_given_pos * n_spam
non_spam_emails
```

```python
# Exercício 3: Decifrando a string "OCZ XJMMZXO VINRZM" com o Processo de Markov

# Função para aplicar a cifra de César com deslocamento
def cifra_cesar(text, shift):
    result = ''
    for char in text:
        if char.isalpha():
            shift_char = ord('A') if char.isupper() else ord('a')
            result += chr((ord(char) - shift_char + shift) % 26 + shift_char)
        else:
            result += char
    return result

# String original
cipher_text = "OCZ XJMMZXO VINRZM"

# Aplicando a cifra de César com deslocamento 7 três vezes
for i in range(3):
    cipher_text = cifra_cesar(cipher_text, 7)
    print(f"Decodificação após {i+1} etapas: {cipher_text}")
```

```python
# Exercício 4: Função de Massa de Probabilidade e Distribuição Cumulativa

# Probabilidades fictícias de um dado lançamento
probabilidades = [0.2, 0.1, 0.25, 0.15, 0.2, 0.1]
resultados = ['1', '2', '3', '4', '5', '6']

# Função de Massa de Probabilidade
plt.bar(resultados, probabilidades)
plt.title('Função de Massa de Probabilidade')
plt.xlabel('Resultados')
plt.ylabel('Probabilidade')
plt.show()

# Distribuição Cumulativa
cumulative_prob = np.cumsum(probabilidades)
plt.plot(resultados, cumulative_prob, marker='o', linestyle='--', color='r')
plt.title('Distribuição Cumulativa')
plt.xlabel('Resultados')
plt.ylabel('Probabilidade Cumulativa')
plt.show()
```

## Capítulo 3: Grafos Básicos

```python
import networkx as nx
import matplotlib.pyplot as plt

# Exercício 1: Desenhar um grafo representando cinco amigos em uma festa

# Criando o grafo com 5 amigos (vértices)
G = nx.Graph()
G.add_nodes_from(["Amigo1", "Amigo2", "Amigo3", "Amigo4", "Amigo5"])

# Adicionando arestas que representam se os amigos se conhecem
arestas = [
    ("Amigo1", "Amigo2"), ("Amigo1", "Amigo3"), ("Amigo1", "Amigo4"),
    ("Amigo2", "Amigo3"), ("Amigo2", "Amigo5"), ("Amigo3", "Amigo4"),
    ("Amigo4", "Amigo5")
]
G.add_edges_from(arestas)

# Desenhando o grafo
nx.draw(G, with_labels=True, node_color='lightblue', node_size=2000, font_size=10)
plt.title('Grafo dos Amigos na Festa')
plt.show()

# Quantas arestas existem?
print(f"Número de arestas: {G.number_of_edges()}")
```

```python
# Exercício 2: Número total de arestas em um grafo completo com n vértices

def total_arestas_grafo_completo(n):
    return (n * (n - 1)) // 2

# Exemplo para n = 5
n = 5
total_arestas = total_arestas_grafo_completo(n)
print(f"Número total de arestas em um grafo completo com {n} vértices: {total_arestas}")
```

```python
# Exercício 3: Determinar se um grafo não direcionado é conexo

def eh_conexo(G):
    return nx.is_connected(G)

# Verificando se o grafo dos amigos é conexo
conexo = eh_conexo(G)
print(f"O grafo é conexo? {conexo}")
```

## Capítulo 4: Grafos Estendidos

```python
# Exercício 1: O que é um grafo bipartido? Desenhe um exemplo

# Um grafo bipartido tem dois conjuntos de vértices, e as arestas só conectam vértices entre os dois conjuntos, não dentro de um conjunto.

# Criando um grafo bipartido
B = nx.Graph()
top_nodes = ['A', 'B', 'C']  # Conjunto 1
bottom_nodes = ['1', '2', '3']  # Conjunto 2
B.add_nodes_from(top_nodes, bipartite=0)
B.add_nodes_from(bottom_nodes, bipartite=1)

# Adicionando arestas entre os conjuntos
B.add_edges_from([('A', '1'), ('A', '2'), ('B', '2'), ('C', '3')])

# Desenhando o grafo bipartido
pos = nx.bipartite_layout(B, top_nodes)
nx.draw(B, pos, with_labels=True, node_color='lightgreen', node_size=2000, font_size=10)
plt.title('Grafo Bipartido')
plt.show()
```

```python
# Exercício 3: Calcular o caminho mais curto em um grafo ponderado

# Criando um grafo ponderado
WG = nx.Graph()
WG.add_weighted_edges_from([
    ('A', 'B', 1.5), ('A', 'C', 2.0), ('B', 'C', 1.0),
    ('B', 'D', 2.5), ('C', 'D', 1.2), ('C', 'E', 2.8), ('D', 'E', 1.5)
])

# Desenhando o grafo ponderado
pos = nx.spring_layout(WG)
nx.draw(WG, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10)
labels = nx.get_edge_attributes(WG, 'weight')
nx.draw_networkx_edge_labels(WG, pos, edge_labels=labels)
plt.title('Grafo Ponderado')
plt.show()

# Calculando o caminho mais curto entre 'A' e 'E'
shortest_path = nx.shortest_path(WG, source='A', target='E', weight='weight')
print(f"Caminho mais curto entre 'A' e 'E': {shortest_path}")
```
